#### 1.8 How to handle null value on the data

```python
df.isnull().sum()
```

```python
df['Arrival Delay in Minutes'] = df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].mean())
```

```python
df.isnull().sum()
```

#### 1.9 How to handle transfrom value on the data

```python
df = df.replace(to_replace=['neutral or dissatisfied','satisfied'], value=[0, 1])
df = df.replace(to_replace=['Female','Male'], value=[0,1])
df = df.replace(to_replace=['disloyal Customer','Loyal Customer'], value=[0,1])
df = df.replace(to_replace=['Personal Travel','Business travel'], value=[0,1])
df = df.replace(to_replace=['Eco','Business','Eco Plus'], value=[0,1,2])
```

```python
for col in df:
    print(f'{col}: {df[col].unique()}')
```

#### 1.10 How to handle outlier on the data

```python
df = df[df['Flight Distance'] < 4000].reset_index(drop=True)

plt.figure(figsize=(3,4))
df['Flight Distance'].plot(kind='box', patch_artist=True);
```

```python
df_to_plot = df.select_dtypes(include=np.number)
df_to_plot.plot(subplots=True, layout=(5,5), kind='box', figsize=(12,14), patch_artist=True)
plt.subplots_adjust(wspace=0.5)
```

#### 1.11 How to handle feature correlation on the data

```python
corrs = df.select_dtypes(include=np.number).corr(method = 'kendall')
sns.clustermap(corrs, cbar_pos=(-0.05, 0.8, 0.05, 0.18), cmap='coolwarm', center=0)
```

#### 2.4 Check how many satisfaction and not satisfaction passengers based on type :

- 2.4.1 Type of Travel:Purpose of the flight of the passengers (Personal Travel, Business Travel)
- 2.4.2 Class:Travel class in the plane of the passengers (Business, Eco, Eco Plus)
- 2.4.3 Customer Type:The customer type (Loyal customer, disloyal customer)

```python
df.head()
```

```python
col_Doc = [ 'Type of Travel', 'Class','Customer Type']
df_flag = df[col_Doc+["satisfaction_v2"]]

length = len(col_Doc)

df_flag["satisfaction_v2"] = df["satisfaction_v2"].replace({0:"Not Satisfaction",1:"Satisfaction"})

fig = plt.figure(figsize=(21,24))

for i,j in itertools.zip_longest(col_Doc,range(length)):
    plt.subplot(5,4,j+1)
    ax = sns.countplot(df_flag[i],hue=df_flag["satisfaction_v2"],palette=["g","r"])
    plt.yticks(fontsize=8)
    plt.xlabel("")
    plt.ylabel("")
    plt.title(i)
```
